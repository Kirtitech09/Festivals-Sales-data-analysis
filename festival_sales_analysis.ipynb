# Exploratory Data Analysis(EDA)

### Gender

```python
plt.figure(figsize = (4,5) )
ax = sns.countplot(x = 'Gender', data = df )
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
gender_wise_sales =  df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Gender' , y = 'Amount' , data = gender_wise_sales )
```

### Age

```python
ax =  sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
for bars in ax.containers : 
    ax.bar_label(bars)
```

```python
Age_Group_wise_sales =  df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Age Group' , y = 'Amount' , data = Age_Group_wise_sales )
```

```python
State_wise_sales =  df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(16,5)})
sns.barplot(x = 'State' , y = 'Orders' , data = State_wise_sales )
```

```python
State_wise_amount =  df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(22,5)})
sns.barplot(x = 'State' , y = 'Amount' , data = State_wise_amount )
```

### Marital Status

```python
plt.figure(figsize = (4,5) )
ax =  sns.countplot(data = df, x = 'Marital_Status')
for bars in ax.containers : 
    ax.bar_label(bars)
```

```python
plt.figure(figsize = (4,5))
marital_sales =  df.groupby(['Marital_Status','Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Marital_Status' , y = 'Amount' , data = marital_sales,hue = 'Gender'  )
```

### Product Category

```python
ax =  sns.countplot(data = df, x = 'Product_Category')
sns.set(rc = {'figure.figsize':(20,5)})
for bars in ax.containers : 
    ax.bar_label(bars)
```

```python
Product_Category_wise_amount =  df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(22,5)})
sns.barplot(x = 'Product_Category' , y = 'Amount' , data = Product_Category_wise_amount )
```

```python
Product_wise_order =  df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(22,5)})
sns.barplot(x = 'Product_ID' , y = 'Orders' , data = Product_wise_order )
```
